```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams  # Импорт параметров конфигурации для настройки шрифтов в Matplotlib

# Настройка параметров шрифта для графиков
rcParams['font.family'] = 'Times New Roman'  # Использование шрифта Times New Roman
rcParams.update({'font.size': 16})  # Установка размера шрифта для текста на графиках

fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # Создаем фигуру и четыре панели для графиков (а, б, в, г)

# Первая панель (а)
x = np.linspace(0.1, 10, 400) # Задаем диапазон значений аргумента х заданных функций
# Заданные функции
f1 = np.sin(x) / x
f2 = -2**(2*x - 1)
# Настройка панели
axs[0, 0].plot(x, f1, label='2) $y=\\frac{\\sin(x)}{x}$', color=(0,1,0), linestyle='-')
axs[0, 0].set_title('a) y=sin(x)/x и y=-2^(2x-1)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend(loc='lower left')
#Добавление второй y-оси на графике для размещения второй функции
ax2 = axs[0,0].twinx()
ax2.plot(x, f2, label='1) $y=-2^{2x-1}$', color=(1,0,0), linestyle='--')
ax2.legend(loc='upper right')
ax2.tick_params(axis='y', labelcolor=(1,0,0))

# Вторая панель (б)
# Задаем начальные данные
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) # сетка из одномерных массивов x и y
Z = np.sin(np.sqrt(X**2 + Y**2))
# Создаем график изолиний
contour = axs[0, 1].contourf(X, Y, Z, levels=20, cmap='viridis')  # levels - количество контурных линий, cmap - цветовая карта
axs[0, 1].set_title('б) $z = \\sin(\\sqrt{x^2 + y^2})$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
colorbar = plt.colorbar(contour, ax=axs[0, 1])
colorbar.set_label('Значение функции')

# Третья панель (в)
 # Оценки за семестр
values = [3, 3, 4, 3, 5, 3, 4]

# Построение гистограммы
axs[1, 0].hist(values, bins=[2.5, 3.5, 4.5, 5.5], edgecolor='black', color='lightblue')
axs[1, 0].set_xlabel('Оценка')
axs[1, 0].set_ylabel('Частота')
axs[1, 0].set_title('в) Гистограмма успеваемости за 1 семестр')

# Четвертая панель (г)
 # Оценки за каждый семестр
values = [4, 3.6, 3, 3.57, 5, 3.7]

# Построение круговой диагараммы
labels = ['1 Семестр', '2 Семестр', '3 Семестр', '4 Семестр', '5 Семестр', '6 Семестр']
axs[1, 1].pie(values, labels=labels, autopct=lambda p: f'{p * sum(values) / 100:.1f}', startangle=140)
axs[1, 1].set_title('г) Круговая диаграмма успеваемости')

# Изменение расположения панелей
plt.tight_layout()

# Отображение графиков
plt.show()
```
