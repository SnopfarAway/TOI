```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams  #        Matplotlib

#     
rcParams['font.family'] = 'Times New Roman'  #   Times New Roman
rcParams.update({'font.size': 16})  #       

fig, axs = plt.subplots(2, 2, figsize=(10, 8)) #        (, , , )

#   ()
x = np.linspace(0.1, 10, 400) #       
#  
f1 = np.sin(x) / x
f2 = -2**(2*x - 1)
#  
axs[0, 0].plot(x, f1, label='2) $y=\\frac{\\sin(x)}{x}$', color=(0,1,0), linestyle='-')
axs[0, 0].set_title('a) y=sin(x)/x  y=-2^(2x-1)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].legend(loc='lower left')
#  y-      
ax2 = axs[0,0].twinx()
ax2.plot(x, f2, label='1) $y=-2^{2x-1}$', color=(1,0,0), linestyle='--')
ax2.legend(loc='upper right')
ax2.tick_params(axis='y', labelcolor=(1,0,0))

#   ()
#   
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) #     x  y
Z = np.sin(np.sqrt(X**2 + Y**2))
#   
contour = axs[0, 1].contourf(X, Y, Z, levels=20, cmap='viridis')  # levels -   , cmap -  
axs[0, 1].set_title(') $z = \\sin(\\sqrt{x^2 + y^2})$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
colorbar = plt.colorbar(contour, ax=axs[0, 1])
colorbar.set_label(' ')

#   ()
 #   
values = [3, 3, 4, 3, 5, 3, 4]

#  
axs[1, 0].hist(values, bins=[2.5, 3.5, 4.5, 5.5], edgecolor='black', color='lightblue')
axs[1, 0].set_xlabel('')
axs[1, 0].set_ylabel('')
axs[1, 0].set_title(')    1 ')

#   ()
 #    
values = [4, 3.6, 3, 3.57, 5, 3.7]

#   
labels = ['1 ', '2 ', '3 ', '4 ', '5 ', '6 ']
axs[1, 1].pie(values, labels=labels, autopct=lambda p: f'{p * sum(values) / 100:.1f}', startangle=140)
axs[1, 1].set_title(')   ')

#   
plt.tight_layout()

#  
plt.show()
```
