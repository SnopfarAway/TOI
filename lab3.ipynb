```python
import pandas as pd
import numpy as np
```

```python
def combine_data(year, months):
  dataset = pd.DataFrame()

  for item in months:
    temp = pd.read_csv(f"Data/basketball/{year}/{item}.csv", parse_dates=True, usecols=[0,1,2,3,4,5,8],skiprows=[0])
    temp.columns=["Date","Start (ET)","Visitor Team","VisitorPts", "Home Team","HomePts","Attend"]
    dataset = pd.concat([dataset, temp])

  return dataset
```

```python
year = 2023
```

```python
#months = ['oct', 'nov', 'dec', 'jan', 'feb', 'mar', 'apr', 'may', 'june'] # 2017,2018,2021,2022
#months = ['oct', 'nov', 'dec', 'jan', 'feb', 'mar', 'jul', 'aug', 'sep', 'oct2'] # 2019
#months = ['dec', 'jan', 'feb', 'mar', 'apr', 'may', 'june', 'jul'] # 2020
months = ['oct', 'nov', 'dec', 'jan', 'feb', 'mar', 'apr'] # 2023
dataset = combine_data(f'{year}', months)
dataset.reset_index(drop=True, inplace=True)
dataset
```

```python
dataset["HomeWin"] = dataset["VisitorPts"] < dataset["HomePts"]
y_true = dataset["HomeWin"].values
dataset
```

```python
dataset["HomeLastWin"] = dataset["HomeWin"]
dataset["VisitorLastWin"] = dataset["VisitorPts"] > dataset["HomePts"]
dataset
```

```python
# заполнить HomeLastWin VisitorLastWin 1 если они прошлую игру выиграли и 0 если проиграли
from collections import defaultdict
won_last = defaultdict(int)

for index, row in dataset.iterrows():
  home_team = row["Home Team"]
  visitor_team = row["Visitor Team"]
  row["HomeLastWin"] = won_last[home_team]
  row["VisitorLastWin"] = won_last[visitor_team]
  dataset.loc[index] = row
  won_last[home_team] = row["HomeWin"]
  won_last[visitor_team] = not row["HomeWin"]

dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
clf = DecisionTreeClassifier(random_state=14)
```

```python
X_previouswins = dataset[["HomeLastWin", "VisitorLastWin"]].values
```

```python
scores = cross_val_score(clf, X_previouswins, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
standings = pd.read_csv(f"Data/basketball/{year}/standings{year-1}.csv", parse_dates=True)
```

```python
# определить у кого из команд в предыдущем сезоне ранг выше
dataset["HomeTeamRanksHigher"] = 0
for index, row in dataset.iterrows():
  home_team = row["Home Team"]
  visitor_team = row["Visitor Team"]

  home_rank = standings[standings["Team"] == home_team]["Rk"].values[0]
  visitor_rank = standings[standings["Team"] == visitor_team]["Rk"].values[0]
  row["HomeTeamRanksHigher"] = int(home_rank > visitor_rank)
  dataset.loc[index] = row
```

```python
X_homehigher = dataset[["HomeLastWin", "VisitorLastWin", "HomeTeamRanksHigher"]].values
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_homehigher, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
# определить кто из команд выиграл предыдущую встречу друг с другом
last_match_winner = defaultdict(int)
dataset["HomeTeamWonLast"] = 0

for index, row in dataset.iterrows():
  home_team = row["Home Team"]
  visitor_team = row["Visitor Team"]
  teams = tuple(sorted([home_team, visitor_team]))
  row["HomeTeamWonLast"] = 1 if last_match_winner[teams] == row["Home Team"] else 0 # проверить кто выиграл встречу HT=1 VT=0
  dataset.loc[index] = row # обновить строку новым значеним
  winner = row["Home Team"] if row["HomeWin"] else row["Visitor Team"] # обновить словарь, чтобы вычислить еще раз когда эти команды встретятся
  last_match_winner[teams] = winner
```

```python
dataset
```

```python
X_lastwinner = dataset[["HomeTeamRanksHigher", "HomeTeamWonLast"]].values
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_lastwinner, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
#преобразовать строковые названия команд к целым числам
from sklearn.preprocessing import LabelEncoder
encoding = LabelEncoder()
encoding.fit(dataset["Home Team"].values)
home_teams = encoding.transform(dataset["Home Team"].values)
visitor_teams = encoding.transform(dataset["Visitor Team"].values)
X_teams = np.vstack([home_teams, visitor_teams]).T
```

```python
# кодировать полученные целочисленные значения в ряд двоичных объектов
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder()

X_teams_expanded = onehot.fit_transform(X_teams)
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_teams_expanded, y_true, scoring='accuracy')
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
X_teams_expanded = onehot.fit_transform(X_teams).todense()
clf.fit(np.asarray(X_teams_expanded), y_true)
dataset["NewHomeWin"] = clf.predict(np.asarray(X_teams_expanded))
```
